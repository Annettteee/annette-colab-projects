```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Define the search parameters
SEARCH_TERMS = ["data science", "computer science", "statistics", "mathematics"]
BASE_URL = "https://annas-archive.org/search?q="

# Store results
books_data = []

headers = {
    "User-Agent": "Mozilla/5.0 (compatible; AnnaScraper/1.0; +https://yourdomain.edu)"
}

for term in SEARCH_TERMS:
    print(f"Searching for books on: {term}")
    search_url = BASE_URL + term.replace(" ", "+")
    response = requests.get(search_url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")

    results = soup.find_all("div", class_="result-item")  # Result box per book

    for item in results:
        try:
            title = item.find("h3").get_text(strip=True)
            link = "https://annas-archive.org" + item.find("a")["href"]
            book_page = requests.get(link, headers=headers)
            book_soup = BeautifulSoup(book_page.content, "html.parser")

            metadata = {
                "Title": title,
                "Authors": None,
                "Year": None,
                "Publisher": None,
                "Tags": None,
                "ISBN": None,
                "Download Link": None,
                "Source Page": link
            }

            # Extract metadata
            info_table = book_soup.find("table")
            if info_table:
                for row in info_table.find_all("tr"):
                    th = row.find("th")
                    td = row.find("td")
                    if not th or not td:
                        continue
                    label = th.text.strip().lower()
                    value = td.text.strip()

                    if "author" in label:
                        metadata["Authors"] = value
                    elif "year" in label:
                        metadata["Year"] = value
                    elif "publisher" in label:
                        metadata["Publisher"] = value
                    elif "isbn" in label:
                        metadata["ISBN"] = value
                    elif "topic" in label or "tag" in label:
                        metadata["Tags"] = value

            # Try to get download link
            download_section = book_soup.find("a", string="Download")
            if download_section:
                metadata["Download Link"] = download_section["href"]

            books_data.append(metadata)

            # Be polite to their server
            time.sleep(2)

        except Exception as e:
            print("Error processing a result:", e)
            continue

# Save results
df = pd.DataFrame(books_data)
df = df.drop_duplicates(subset="Title")
df.to_csv("data_science_books_2020_2025.csv", index=False)

print("‚úÖ Done! Saved results to CSV.")
```

```python
from bs4 import BeautifulSoup
import requests

def get_book_details(work_url):
    try:
        res = requests.get(work_url)
        soup = BeautifulSoup(res.text, 'html.parser')

        # Get description from meta tag or fallback
        description = ""
        meta_desc = soup.find("meta", attrs={"name": "description"})
        if meta_desc:
            description = meta_desc["content"]
        else:
            # Fallback: try OpenLibrary's description container
            desc_div = soup.find("div", class_="book-description")
            if desc_div:
                description = desc_div.get_text(strip=True)

        # Get actual read/borrow/download link
        read_link = None
        read_btn = soup.find("a", string=lambda t: t and ("Read" in t or "Borrow" in t or "Preview" in t))
        if read_btn and read_btn.has_attr("href"):
            read_link = "https://openlibrary.org" + read_btn["href"]

        return pd.Series([description, read_link])

    except Exception as e:
        return pd.Series([None, None])

# Apply to all books (feel free to slice for testing)
df[['Description', 'Book Access Link']] = df['OpenLibrary URL'].apply(get_book_details)

# Save to CSV
df.to_csv("Books_With_Descriptions_And_Links.csv", index=False)
df[['Title', 'Description', 'Book Access Link']].head()
```

```python
import requests
import pandas as pd
from tqdm.notebook import tqdm

def get_better_details(work_url):
    try:
        work_id = work_url.split('/')[-1]
        work_api = f"https://openlibrary.org/works/{work_id}.json"
        edition_api = f"https://openlibrary.org/works/{work_id}/editions.json?limit=1"

        # Step 1: Get the main work metadata
        desc = ""
        res = requests.get(work_api)
        if res.status_code == 200:
            data = res.json()
            if isinstance(data.get("description"), dict):
                desc = data["description"].get("value", "")
            elif isinstance(data.get("description"), str):
                desc = data["description"]

        # Step 2: Try to find a readable edition
        link = None
        res_ed = requests.get(edition_api)
        if res_ed.status_code == 200:
            eds = res_ed.json()
            if eds.get("entries"):
                first_edition = eds["entries"][0]
                edition_key = first_edition.get("key")
                if edition_key:
                    link = "https://openlibrary.org" + edition_key

        return pd.Series([desc, link])

    except Exception as e:
        return pd.Series([None, None])

# Add progress bar
tqdm.pandas()
sample = df.head(100)  # You can increase this to 500+ later
sample[['Summary/Abstract', 'Actual Book Link']] = sample['OpenLibrary URL'].progress_apply(get_better_details)

# Save
sample.to_csv("Books_With_Abstracts_And_Links.csv", index=False)
sample[['Title', 'Summary/Abstract', 'Actual Book Link']].head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

BASE_URL = "https://link.springer.com"
search_url = (
    "https://link.springer.com/search?facet-content-type=Book&facet-language=%22En%22"
    "&package=openaccess&query=data+science+mathematics+statistics+computer+science"
)

headers = {
    "User-Agent": "Mozilla/5.0"
}

books = []

# Iterate through first 5 pages
for page in range(1, 6):
    print(f"Scraping page {page}...")
    page_url = f"{search_url}&page={page}"
    response = requests.get(page_url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    for item in soup.select('.content-item-list .content-item'):
        try:
            title_elem = item.select_one('h2 a')
            title = title_elem.text.strip()
            book_url = BASE_URL + title_elem['href']

            # Follow book page to get more info
            book_resp = requests.get(book_url, headers=headers)
            book_soup = BeautifulSoup(book_resp.text, 'html.parser')

            # Author
            author = book_soup.select_one('.authors__name') or book_soup.select_one('.authors__list')
            author = author.text.strip() if author else "N/A"

            # Abstract / Summary
            abstract_elem = book_soup.select_one('section.Abstract p')
            abstract = abstract_elem.text.strip() if abstract_elem else "No abstract available."

            # Year
            year_elem = book_soup.select_one('span[itemprop="datePublished"]')
            year = int(year_elem.text.strip()) if year_elem else "Unknown"
            if year < 2020 or year > 2025:
                continue

            # PDF link
            pdf_link_tag = book_soup.select_one('a[title="Download this book in PDF format"]')
            pdf_link = BASE_URL + pdf_link_tag['href'] if pdf_link_tag else "Not available"

            books.append({
                'Title': title,
                'Author': author,
                'Abstract': abstract,
                'Year': year,
                'PDF_Link': pdf_link
            })

            time.sleep(1)  # Be polite with server
        except Exception as e:
            print(f"Skipping a book due to error: {e}")
            continue

# Convert to DataFrame
df = pd.DataFrame(books)

# Save to CSV
df.to_csv("Open_Access_Books_2020_2025.csv", index=False)
print("Done. CSV saved as Open_Access_Books_2020_2025.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Load your existing CSV (replace with your path)
df = pd.read_csv('/content/All_Books_2020_2025.csv')

# Add columns for real book URL and description
df['Actual_Book_Link'] = ''
df['Description'] = ''

headers = {"User-Agent": "Mozilla/5.0"}

for idx, row in df.iterrows():
    try:
        url = row['OpenLibrary URL']
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract readable link (archive.org) ‚Äî might be in <a class="cta-button" href="...">
        read_link_tag = soup.select_one('a.cta-button')
        read_link = 'https://openlibrary.org' + read_link_tag['href'] if read_link_tag else 'Not available'

        # Extract description
        desc_tag = soup.select_one('#description span, #description p')
        description = desc_tag.text.strip() if desc_tag else "No description available."

        df.at[idx, 'Actual_Book_Link'] = read_link
        df.at[idx, 'Description'] = description

        print(f"‚úì Processed: {row['Title']}")
        time.sleep(1)
    except Exception as e:
        print(f"‚úó Failed on: {row['Title']} ‚Äî {e}")
        continue

# Save the updated CSV
df.to_csv("Books_with_Actual_Links.csv", index=False)
print("‚úÖ CSV saved as 'Books_with_Actual_Links.csv'")
```
